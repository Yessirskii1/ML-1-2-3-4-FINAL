## Визуализация распределения переменных

```python
#числовые переменные

plt.figure(figsize=(16, 12))

# Salary Per Month
plt.subplot(3, 2, 1)
sns.histplot(df['Salary Per Month'], kde=True)
plt.title('Распределение месячной зарплаты')

# Working Hours
plt.subplot(3, 2, 2)
sns.histplot(df['Working Hours'], kde=True)
plt.title('Распределение рабочих часов')

# Package
plt.subplot(3, 2, 3)
sns.histplot(df['Package'], kde=True)
plt.title('Распределение годового пакета')

# Age
plt.subplot(3, 2, 4)
sns.histplot(df['age'], kde=True)
plt.title('Распределение возраста')

# OverTime Pay
plt.subplot(3, 2, 5)
sns.histplot(df['OverTime Pay'], kde=True)
plt.title('Распределение оплаты сверхурочных')

plt.tight_layout()
plt.show()
```

```python
#категориальные  переменные

num_cols = df.select_dtypes(include='object').columns # или object
fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))

for ax, col in zip(axes, num_cols):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```
